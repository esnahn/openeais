## 전국 지하층 주택 규모 분석

주택 정의:

1. 주 용도 코드가 01 또는 02로 시작하고, 기타 용도에 주택, 다세대, 다가구, 공관, 아파트, 기숙사 등의 텍스트가 포함됨
2. 주 용도 코드가 01 또는 02로 시작하고, 기타 용도에 주택, 다세대, 다가구, 공관, 아파트, 기숙사 등의 텍스트가 포함되며, 설비, 배관, 창고, 계단, 승강기, 경비, 주차, 보관, 경로당, 보육, 문고, 독서, 기계, 전기, 체육, 사무소, MDF, 방재 등의 텍스트가 포함되지 않음

과대추정이 될 방식과 과소추정이 될 방식 두 가지로 주택 규모를 산출하여, 최종 결과의 오차 범위를 가늠함

지하층 정의: 층 구분 코드가 10임

반지하(층 바닥이 지면보다 낮지만 천장은 지면보다 높음) 추정은 불가능. 각 층의 바닥 및 천장 표고 (또는, 최고/최저 표고)가 건축물대장 또는 건축허브 등에 구축되어야 함.

반드시 지하 1층이 지면과 가까운 층인 것도 아님. 부산 등의 경우 지면에서 수평으로 지하 2층에 진입할 수 있는 경우 있음.

```python
housing_contains = "|".join(["주택", "다세대", "다가구", "공관", "아파트", "기숙사"])
no_housing_contains = "|".join(
    [
        "설비",
        "배관",
        "창고",
        "계단",
        "승강기",
        "경비",
        "주차",
        "보관",
        "경로당",
        "보육",
        "문고",
        "독서",
        "기계",
        "전기",
        "체육",
        "사무소",
        "MDF",
        "방재",
    ]
)

ddf1 = ddf[
    (ddf["층_구분_코드"] == "10")
    & (ddf["주_용도_코드"].str.contains("^01|^02"))
    & (ddf["기타_용도"].str.contains(housing_contains))
]
ddf2 = ddf1[~ddf1["기타_용도"].str.contains(no_housing_contains)]
```

```python
with ProgressBar():
    ddf1.to_parquet("underground_housing1")
    ddf2.to_parquet("underground_housing2")
```

### 3번 정의

주택 정의 3. 주 용도 코드가 단독주택, 다중주택, 다가구주택, 연립주택, 다세대주택 중 하나이고, 기타 용도에 주택, 다세대, 다가구 등의 텍스트가 포함되며, 설비, 배관, 창고, 계단, 승강기, 경비, 주차, 보관, 경로당, 보육, 문고, 독서, 기계, 전기, 체육, 사무소, MDF, 방재 등의 텍스트가 포함되지 않음

아파트를 제외한 수치. (아파트는 반지하가 없다는 판단 또는 과소추정 유도)

```python
# 단독, 다중, 다가구, 연립, 다세대만
narrower_housing = "|".join(
    [
        "01001",
        "01002",
        "01003",
        "02002",
        "02003",
    ]
)
narrower_housing_contains = "|".join(["주택", "다세대", "다가구"])

ddf3 = ddf[
    (ddf["층_구분_코드"] == "10")
    & (ddf["주_용도_코드"].str.contains(narrower_housing))
    & (ddf["기타_용도"].str.contains(narrower_housing_contains))
    & (~ddf1["기타_용도"].str.contains(no_housing_contains))
]

with ProgressBar():
    ddf3.to_parquet("underground_housing3")
```
