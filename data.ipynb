```python
import pandas as pd
import dask.dataframe as dd
```

```python
# 파일 형식 확인
# 층별개요 압축 해제하여 data 폴더 안에 mart_djy_04.txt 파일 준비

# read 1 row as raw text
with open("data/mart_djy_04.txt", "r", encoding="cp949") as t:
    print(t.readline())
```

```python
# compression: str or dict, default ‘infer’
# For on-the-fly decompression of on-disk data.
# If ‘infer’ and ‘%s’ is path-like, then detect compression from the following extensions:
# ‘.gz’, ‘.bz2’, ‘.zip’, ‘.xz’, or ‘.zst’ (otherwise no compression).
# If using ‘zip’, the ZIP file must contain only one data file to be read in.
# https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

# dask doesn't work with zip smoothly. just use txt file.

import openeais

# read 10 rows with pandas
df = pd.read_csv(
    "data/mart_djy_04.txt",
    encoding="cp949",
    header=None,
    sep="|",
    names=openeais.floor_dtypes.keys(),
    dtype=openeais.floor_dtypes,
    nrows=10,  # read only 10 rows; you can't read a really big file with pandas
)
df
```

```python
# https://docs.dask.org/en/stable/generated/dask.dataframe.read_csv.html

ddf = dd.read_csv(
    "data/mart_djy_04.txt",
    # blocksize=25e6    # Default value is computed based on available physical
    #                   # memory and the number of cores, up to a maximum of 64MB.
    encoding="cp949",
    header=None,
    sep="|",
    names=openeais.floor_dtypes.keys(),
    dtype=openeais.floor_dtypes,
)
ddf
```

```python
from dask.diagnostics import ProgressBar

with ProgressBar():
    # ddf = ddf.set_index("관리_건축물대장_PK")
    ddf.to_parquet("test/floor")
```
