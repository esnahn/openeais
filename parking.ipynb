```python
### main part

# hierarchy

df_master_building = df_master.select(
    [
        "관리_상위_건축물대장_PK",
        "대장_구분_코드",
        "대장_구분_코드_명",
        "대장_종류_코드",
        "대장_종류_코드_명",
        "대지_위치",
        "도로명_대지_위치",
        "건물_명",
        "시군구_코드",
        "법정동_코드",
        "대지_구분_코드",
        "번",
        "지",
        "관리_건축물대장_PK",
    ]
).filter(
    (pl.col("대장_종류_코드").is_in(["2", "3"]) & pl.col("관리_상위_건축물대장_PK").is_not_null())
)
# df_master_building.head().collect()

# building usage code

df_use = df_floor.select(
    [
        "시군구_코드",
        # "주_용도_코드",
        # "주_용도_코드_명",
        "관리_건축물대장_PK",
    ]
)

df_apt = df_use.filter(pl.col("주_용도_코드") == "02001").groupby("관리_건축물대장_PK").first()
df_apt = df_apt.join(
    df_master_building.select(
        [
            "관리_건축물대장_PK",
            "관리_상위_건축물대장_PK",
        ]
    ),
    "관리_건축물대장_PK",
    "left",
)
df_apt_complex = df_apt.filter(pl.col("관리_상위_건축물대장_PK").is_not_null())
df_apt_single = df_apt.filter(pl.col("관리_상위_건축물대장_PK").is_null())

df_evbuilding = (
    df_use.filter(pl.col("주_용도_코드").is_in(ev_use_excl_apt))
    .groupby("관리_건축물대장_PK")
    .first()
    .join(
        df_master_building.select(
            [
                "관리_건축물대장_PK",
                "관리_상위_건축물대장_PK",
            ]
        ),
        "관리_건축물대장_PK",
        "left",
    )
)
df_evbuilding_complex = df_evbuilding.filter(pl.col("관리_상위_건축물대장_PK").is_not_null())
df_evbuilding_single = df_evbuilding.filter(pl.col("관리_상위_건축물대장_PK").is_null())
```

```python
df_apt_complex_site = df_apt_complex.join(df_dong.select(col_dong), "관리_건축물대장_PK")
df_apt_complex_site = df_apt_complex_site.groupby("관리_상위_건축물대장_PK").agg(
    [
        pl.first("시군구_코드"),
        # pl.first("주_용도_코드"),
        # pl.first("주_용도_코드_명"),
        pl.sum("세대_수(세대)"),
        pl.sum("옥내_자주식_대수(대)"),
        pl.sum("옥외_자주식_대수(대)"),
        pl.max("사용승인_일"),
    ]
)
df_apt_complex_site = df_apt_complex_site.join(
    df_site.select(col_site),
    left_on="관리_상위_건축물대장_PK",
    right_on="관리_건축물대장_PK",
    suffix="_site",
)
df_apt_complex_site = df_apt_complex_site.filter((pl.col("주_용도_코드") == "02000"))

# 100 units cut

df_apt_complex_site = df_apt_complex_site.filter(
    (pl.col("세대_수(세대)_site") >= 100)
    | ((pl.col("세대_수(세대)_site") == 0) & (pl.col("세대_수(세대)") >= 100))
)


df_apt_complex_site = df_apt_complex_site.with_columns(
    [
        (
            pl.when(pl.col("옥내_자주식_대수(대)_site") + pl.col("옥외_자주식_대수(대)_site") > 0)
            .then(pl.col("옥내_자주식_대수(대)_site") + pl.col("옥외_자주식_대수(대)_site"))
            .otherwise(pl.col("옥내_자주식_대수(대)") + pl.col("옥외_자주식_대수(대)"))
            .alias("옥내외_자주식_대수")
        )
    ]
)

df_apt_complex_site = df_apt_complex_site.filter(pl.col("옥내외_자주식_대수") >= 50)
df_apt_complex_site = df_apt_complex_site.with_columns(
    (pl.col("옥내외_자주식_대수") * 0.02).ceil().cast(pl.Int64).alias("전기차_기축_대수")
)
df_apt_complex_site = df_apt_complex_site.with_columns(
    pl.when((pl.col("사용승인_일") >= "20130000"))
    .then((pl.col("옥내외_자주식_대수") * 0.05).ceil())
    .otherwise(0)
    .cast(pl.Int64)
    .alias("전기차_신축_최근10년_대수")
)


df_apt_complex_result = df_apt_complex_site.collect()
df_apt_complex_result
```

```python
### 부설주차장 excluding apt.

df_evbuilding_complex_site = df_evbuilding_complex.join(
    df_dong.select(col_dong), "관리_건축물대장_PK"
)

# filter apt.
df_evbuilding_complex_site = df_evbuilding_complex_site.join(
    df_apt_complex_result.lazy().select(["관리_상위_건축물대장_PK", "시군구_코드"]),
    "관리_상위_건축물대장_PK",
    how="left",
    suffix="_apt",
).filter(pl.col("시군구_코드_apt").is_null())

df_evbuilding_complex_site = df_evbuilding_complex_site.groupby("관리_상위_건축물대장_PK").agg(
    [
        pl.first("시군구_코드"),
        # pl.first("주_용도_코드"),
        # pl.first("주_용도_코드_명"),
        pl.sum("세대_수(세대)"),
        pl.sum("옥내_자주식_대수(대)"),
        pl.sum("옥외_자주식_대수(대)"),
        pl.max("사용승인_일"),
    ]
)
df_evbuilding_complex_site = df_evbuilding_complex_site.join(
    df_site.select(col_site),
    left_on="관리_상위_건축물대장_PK",
    right_on="관리_건축물대장_PK",
    suffix="_site",
)

# filter no ev use
df_evbuilding_complex_site = df_evbuilding_complex_site.filter(
    ~(pl.col("주_용도_코드").is_in(no_ev_use))
)

df_evbuilding_complex_site = df_evbuilding_complex_site.with_columns(
    [
        (
            pl.when(pl.col("옥내_자주식_대수(대)_site") + pl.col("옥외_자주식_대수(대)_site") > 0)
            .then(pl.col("옥내_자주식_대수(대)_site") + pl.col("옥외_자주식_대수(대)_site"))
            .otherwise(pl.col("옥내_자주식_대수(대)") + pl.col("옥외_자주식_대수(대)"))
            .alias("옥내외_자주식_대수")
        )
    ]
)
df_evbuilding_complex_site = df_evbuilding_complex_site.filter(
    pl.col("옥내외_자주식_대수") >= 50
)
df_evbuilding_complex_site = df_evbuilding_complex_site.with_columns(
    (pl.col("옥내외_자주식_대수") * 0.02).ceil().cast(pl.Int64).alias("전기차_기축_대수")
)
df_evbuilding_complex_site = df_evbuilding_complex_site.with_columns(
    pl.when((pl.col("사용승인_일") >= "20130000"))
    .then((pl.col("옥내외_자주식_대수") * 0.05).ceil())
    .otherwise(0)
    .cast(pl.Int64)
    .alias("전기차_신축_최근10년_대수")
)

df_evbuilding_complex_result = df_evbuilding_complex_site.collect()
df_evbuilding_complex_result
```
